```python
import json
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# File path
filepath = 'data/yelp_academic_dataset_business.json'
```

```python
# Read JSON file and extract raw data
raw_data_df = pd.read_json(filepath, lines=True)
raw_data_df.head()
```

```python
# Make a copy and drop unneeded columns
copy_df = raw_data_df[['name', 'address', 'city', 'state', 'postal_code', 'latitude', 'longitude', 'stars', 'review_count', 'categories']].copy().dropna()
copy_df.head()
```

```python
# Filter for 'Restaurants' in 'Porltand'
criteria_df = copy_df[(copy_df['categories'].str.contains('Restaurants')) & (copy_df['city'] == 'Portland')]
criteria_df.head()
```

```python
# Reset the index
clean_df = criteria_df.reset_index(drop=True)
clean_df
```

```python
# Create the sqlite db
path = 'data/portland.sqlite'
engine = create_engine(f'sqlite:///{path}', echo=True)
```

```python
# Open a connection
conn = engine.connect()
```

```python
# Save the dataframe to the sqlite db
table = 'restaurants'
clean_df.to_sql(table, conn, index_label='id', if_exists='fail')
```

```python
# Create a session (link) from Python to the DB
session = Session(engine)
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)
```

```python
# Using the inspector to print the column names within the 'restaurants' table and its types
columns = inspector.get_columns(table)
for column in columns:
    print(column['name'], column['type'])
```

```python
# Query all records in the the database table
data = pd.read_sql('SELECT * FROM restaurants', conn)
```
